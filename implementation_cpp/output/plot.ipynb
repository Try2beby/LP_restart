```python
import numpy as np
import matplotlib.pyplot as plt

# read file
A_t = np.loadtxt('./trust_obj_residual.txt')
A_g = np.loadtxt('./gurobi_obj_residual.txt')

print(A_t.shape,A_g.shape)
# check if A[:,2] is all 2
print(np.unique(A_g[:,2]))
```

```python
%matplotlib inline
from matplotlib_inline import backend_inline

backend_inline.set_matplotlib_formats("svg")

# set dpi
plt.figure(dpi=600)
plt.plot(A_t[:,0]-A_g[:,0])
plt.title("trust obj minus gurobi obj")
plt.savefig("trust_obj_minus_gurobi_obj.png")
```

```python
plt.figure(dpi=600)
plt.plot(A_g[:,1])
plt.title("gurobi residual: ||z-z0||-r")
plt.savefig("gurobi_residual.png")
```
